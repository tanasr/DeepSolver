```python
# define ode function
def func(x, t):
    dydx = 32 - 0.16 * x
    return dydx

# def func(x, t):
#     dydx = np.exp(2*x)
#     return dydx

x0, xN, h = 0, 30, 0.1
x = np.arange(x0, xN+h, h)
ymin, ymax, ystep = 0, 400, 100
y = np.arange(ymin, ymax+ystep, ystep)

# compute solutions for different boundary values
for y0 in y:
    sol = odeint(func, y0, x)
    plt.plot(x, sol, label='y0 = ' + str(y0))
    # break

xgrid = np.linspace(x0, xN, 20)
X, Y = np.meshgrid(xgrid, y)

U = 1 #unit length for vectors
V = func(Y, None) #vector
N = np.sqrt(U**2 + V**2)
U /= N
V /= N

plt.figure
plt.quiver(X, Y, U, V, angles='xy')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.axis([x0, xN, ymin, ymax])
plt.legend()
plt.title(r'Solution of $y^\prime(x) = 32-0.16x$', fontsize='large')
plt.show()
```
